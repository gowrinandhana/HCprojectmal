```python
# Install required libraries
!pip install SpeechRecognition pydub
!apt install ffmpeg

import os
import sys
import speech_recognition as sr
from pydub import AudioSegment
from google.colab import files

# Upload the .ogg file (run this cell first)
uploaded = files.upload()
input_file = next(iter(uploaded))  # Get the uploaded filename

# Define paths
orig_song = input_file
dest_song = "result.wav"

# Convert .ogg to .wav
def convert_ogg_to_wav():
    song = AudioSegment.from_ogg(orig_song)
    song.export(dest_song, format="wav")

convert_ogg_to_wav()

# Recognize speech from the .wav file
r = sr.Recognizer()
with sr.AudioFile(dest_song) as source:
    audio = r.record(source)  # Read the entire audio file

    try:
        text = r.recognize_google(audio, language='ml-IN')  # Malayalam language
        print("Recognized Text:")
        print(text)
    except sr.UnknownValueError:
        print("Google Speech Recognition could not understand audio")
    except sr.RequestError as e:
        print(f"Could not request results from Google Speech Recognition; {e}")
```

```python
from IPython.display import HTML, Javascript, display
import time

# HTML/JavaScript interface for real-time speech recognition
display(HTML('''
<div>
  <button id="startBtn" style="padding:10px;background:#4CAF50;color:white;border:none;border-radius:4px;">Start Listening</button>
  <button id="stopBtn" style="padding:10px;background:#f44336;color:white;border:none;border-radius:4px;" disabled>Stop Listening</button>
  <div id="output" style="margin-top:20px;padding:10px;border:1px solid #ddd;min-height:100px;"></div>
</div>

<script>
const outputDiv = document.getElementById('output');
const startBtn = document.getElementById('startBtn');
const stopBtn = document.getElementById('stopBtn');
let recognition;

// Check if browser supports speech recognition
if (!('webkitSpeechRecognition' in window)) {
  outputDiv.innerHTML = '<p style="color:red;">Your browser doesn\'t support speech recognition. Try Chrome.</p>';
  startBtn.disabled = true;
} else {
  recognition = new webkitSpeechRecognition();
  recognition.continuous = true;
  recognition.interimResults = true;
  recognition.lang = 'ml-IN'; // Malayalam language code

  recognition.onresult = function(event) {
    let finalTranscript = '';
    let interimTranscript = '';

    for (let i = event.resultIndex; i < event.results.length; i++) {
      const transcript = event.results[i][0].transcript;
      if (event.results[i].isFinal) {
        finalTranscript += transcript;
      } else {
        interimTranscript += transcript;
      }
    }

    // Send results to Python
    if (finalTranscript) {
      google.colab.kernel.invokeFunction('processResult', [finalTranscript], {});
    }
  };

  recognition.onerror = function(event) {
    outputDiv.innerHTML += '<p style="color:red;">Error: ' + event.error + '</p>';
  };

  startBtn.onclick = function() {
    outputDiv.innerHTML = '<p>Listening... Speak now (Malayalam)</p>';
    startBtn.disabled = true;
    stopBtn.disabled = false;
    recognition.start();
  };

  stopBtn.onclick = function() {
    recognition.stop();
    startBtn.disabled = false;
    stopBtn.disabled = true;
    outputDiv.innerHTML += '<p>Stopped listening</p>';
  };
}
</script>
'''))

# Function to process results in Python
def process_result(result):
    print("Recognized Malayalam Text:")
    print(result)
    return result

# Register the callback function
from google.colab import output
output.register_callback('processResult', process_result)
```

```python
from IPython.display import HTML, Javascript, display
import time

# First, request microphone permission
display(Javascript('''
  async function requestMic() {
    try {
      const stream = await navigator.mediaDevices.getUserMedia({audio: true});
      console.log('Microphone access granted');
      return true;
    } catch (err) {
      console.error('Microphone access error:', err);
      return false;
    }
  }
  requestMic();
'''))

# HTML/JavaScript interface with improved error handling
display(HTML('''
<div>
  <button id="startBtn" style="padding:10px;background:#4CAF50;color:white;border:none;border-radius:4px;margin:5px;">Start Listening</button>
  <button id="stopBtn" style="padding:10px;background:#f44336;color:white;border:none;border-radius:4px;margin:5px;" disabled>Stop Listening</button>
  <div id="status" style="margin:10px 0;color:#666;"></div>
  <div id="output" style="margin-top:20px;padding:10px;border:1px solid #ddd;min-height:100px;background:#f9f9f9;"></div>
</div>

<script>
const outputDiv = document.getElementById('output');
const statusDiv = document.getElementById('status');
const startBtn = document.getElementById('startBtn');
const stopBtn = document.getElementById('stopBtn');
let recognition;

function updateStatus(message, isError = false) {
  statusDiv.innerHTML = isError ? `<span style="color:red">${message}</span>` : message;
}

// Check browser support
if (!('webkitSpeechRecognition' in window)) {
  updateStatus("Your browser doesn't support speech recognition. Please use Chrome or Edge.", true);
  startBtn.disabled = true;
} else {
  recognition = new webkitSpeechRecognition();
  recognition.continuous = true;
  recognition.interimResults = true;
  recognition.lang = 'ml-IN'; // Malayalam language code

  recognition.onstart = function() {
    updateStatus("Listening... Speak in Malayalam");
    outputDiv.innerHTML = '';
  };

  recognition.onresult = function(event) {
    let finalTranscript = '';

    for (let i = event.resultIndex; i < event.results.length; i++) {
      if (event.results[i].isFinal) {
        finalTranscript += event.results[i][0].transcript;
      }
    }

    if (finalTranscript) {
      outputDiv.innerHTML += '<p>' + finalTranscript + '</p>';
      try {
        google.colab.kernel.invokeFunction('processResult', [finalTranscript], {});
      } catch (e) {
        console.error('Error sending to Python:', e);
      }
    }
  };

  recognition.onerror = function(event) {
    updateStatus("Error: " + event.error, true);
    startBtn.disabled = false;
    stopBtn.disabled = true;
  };

  recognition.onend = function() {
    updateStatus("Ready to listen");
    startBtn.disabled = false;
    stopBtn.disabled = true;
  };

  startBtn.onclick = async function() {
    try {
      const hasPermission = await navigator.permissions.query({name: 'microphone'});
      if (hasPermission.state !== 'granted') {
        updateStatus("Please allow microphone access first", true);
        return;
      }

      recognition.start();
      startBtn.disabled = true;
      stopBtn.disabled = false;
    } catch (err) {
      updateStatus("Error starting recognition: " + err.message, true);
    }
  };

  stopBtn.onclick = function() {
    recognition.stop();
  };
}
</script>
'''))

# Function to process results in Python
def process_result(result):
    print("Recognized Malayalam Text:", result)
    return result

# Register the callback function
from google.colab import output
output.register_callback('processResult', process_result)

print("Ready for Malayalam speech recognition. Click 'Start Listening' and speak.")
```
